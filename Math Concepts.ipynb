# Math Concepts

## Properties or laws of logarithms:

---

### Fundamental Logarithm Rules

- **Product Rule:**  
  $$
  \log_b(mn) = \log_b m + \log_b n
  $$  
  The logarithm of a product equals the sum of the logarithms.

- **Quotient Rule:**  
  $$
  \log_b\left(\frac{m}{n}\right) = \log_b m - \log_b n
  $$  
  The logarithm of a quotient equals the difference of the logarithms.

- **Power Rule:**  
  $$
  \log_b(m^n) = n \log_b m
  $$  
  The logarithm of a number raised to a power equals the exponent times the logarithm of the base number.

- **Root Rule:**  
  $$
  \log_b\left(\sqrt[n]{m}\right) = \frac{1}{n} \log_b m
  $$  
  The logarithm of an n-th root is the logarithm divided by n.

- **Change of Base Rule:**  
  $$
  \log_b a = \frac{\log_c a}{\log_c b}
  $$  
  Allows you to compute logarithms with any base using another base (commonly base 10 or $$e$$).

- **Base Switch Rule (Reciprocal Rule):**  
  $$
  \log_b a = \frac{1}{\log_a b}
  $$  
  The logarithm base $$b$$ of $$a$$ is the reciprocal of the logarithm base $$a$$ of $$b$$.

---

### Special Logarithm Values

- **Logarithm of 1:**  
  $$
  \log_b 1 = 0
  $$  
  Any base logarithm of 1 is 0.

- **Logarithm of the Base:**  
  $$
  \log_b b = 1
  $$  
  Any base logarithm of itself is 1.

- **Inverse Properties:**  
  $$
  b^{\log_b x} = x
  $$  
  $$
  \log_b (b^x) = x
  $$  
  Exponential and logarithmic functions "undo" each other.

---

### Additional Properties

- **Undefined at Zero:**  
  $$
  \log_b 0
  $$  
  is undefined[5][6].

- **Logarithm of Infinity:**  
  $$
  \lim_{x \to \infty} \log_b x = \infty
  $$  
  The logarithm grows without bound as its argument increases.
