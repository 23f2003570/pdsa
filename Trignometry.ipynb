```python
# %load utils/measure.py
import time
from functools import wraps

def measure(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter_ns()
        result = func(*args, **kwargs)
        end = time.perf_counter_ns()
        elapsed_ns = end - start
        
        if elapsed_ns < 1_000:
            time_str = f"{elapsed_ns} ns"
        elif elapsed_ns < 1_000_000:
            time_str = f"{elapsed_ns / 1_000:.3f} Âµs"
        elif elapsed_ns < 1_000_000_000:
            time_str = f"{elapsed_ns / 1_000_000:.3f} ms"
        else:
            time_str = f"{elapsed_ns / 1_000_000_000:.3f} s"
        
        print(f"Performance: {func.__name__}: {time_str}")
        return result
    print("measure-new (util) loaded into global scope.")
    return wrapper
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01) # start, stop, step [start, stop)
s = 1 + np.cos(2 * np.pi * t)
fig, ax = plt.subplots()
ax.plot(t,s)

plt.show()
```

## Polar Coordinates

$(r, \theta)$ instead of $(x,y)$

$r = \sqrt{x^2 + y^2}$

$\theta = \tan^{-1}{\dfrac{y}{x}}$

```python
class Polar:
    def __init__(self):
        pass
        
    def get_polar(self, x, y):
        r = math.sqrt(x**2 + y**2)
        theta = math.atan(y/x)
        print(f"Polar version of point ({x},{y}) is ({r}, {theta}rad)")
        return (r, theta)
    def polar_to_cartesian(self, r, theta):
        x = math.cos(theta) * r
        y = math.sin(theta) * r
        print(f"Cartesian version of ({r}, {theta}rad) is ({x},{y})")
        return (x,y)

p = Polar().get_polar(10,20)
c = Polar().polar_to_cartesian(p[0], p[1])
```
