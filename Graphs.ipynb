## Adjacency Matrix

```python
edges = [(0,1),(1,2),(2,0),(4,0),(0,4),(4,3),(3,4),(3,6),(5,3),(6,5),(5,7),(8,5),(7,8),(8,9),(9,8),(4,7),(7,4)]

import numpy as np
A = np.zeros(shape=(10,10))
for (r,c) in edges:
    A[r,c] = 1
print(A)
```

## Undirected Adjacency Matrix

```python
edges = [(0,1),(1,2),(2,0),(4,0),(0,4),(4,3),(3,4),(3,6),(5,3),(6,5),(5,7),(8,5),(7,8),(8,9),(9,8),(4,7),(7,4)]

import numpy as np
A = np.zeros(shape=(10,10))
for (r,c) in edges:
    A[r,c] = A[c,r] = 1
print(A)
```

Note the matrix is symmetric on the diagonal
Degree of a vertex is the number of incudental columns
In a directed graph, the row indicates the outdegree, the column shows the indegree
In the Adjacency Matrix, the row number or column number indicate the vertex

## Adjacency List

```python
edges = [(0,1),(1,2),(2,0),(4,0),(0,4),(4,3),(3,4),(3,6),(5,3),(6,5),(5,7),(8,5),(7,8),(8,9),(9,8),(4,7),(7,4)]


A = {}
for (r,c) in edges:
    if r in A:
        A[r].append(c)
    else:
        A[r] = [c]

print(A)
```
